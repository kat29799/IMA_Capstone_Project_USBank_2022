# Credit Score <=550 

Chris said we don't have to worry about this one because the data would not be suffiecient as most people can't get a mortgage in this region

```python
CS_A = pd.read_csv('./OTM550.csv')
CS_A.head()
```

```python
CS_A.rename(columns = {'Row Labels':'Date'}, inplace = True)
CS_A.head()
```

```python
CS_A.info()
```

```python
# Make dataframes for each incentive class
# 1 is >-0.75 and <=-0.5
# 2 is >-1 and <=-0.75
# 3 is >-1.25 and <=-1
CS_A1 = CS_A[['>-0.75 and <=-0.5']]
CS_A2 = CS_A[['>-1 and <=-0.75']]
CS_A3 = CS_A[['>-1.25 and <=-1']]
```

```python
CS_A_pltData = [CS_A1.to_numpy().reshape(len(CS_A1.to_numpy())), CS_A2.to_numpy().reshape(len(CS_A2.to_numpy())), CS_A3.to_numpy().reshape(len(CS_A3.to_numpy()))]
```

```python
fig = plt.figure(figsize =(10, 7))
ax = fig.add_subplot(111)
 
# Creating axes instance
bp = ax.boxplot(CS_A_pltData, patch_artist = True, vert = 0)

# x-axis labels
ax.set_yticklabels(['>-0.75 and <=-0.5', '>-1 and <=-0.75',
                    '>-1.25 and <=-1'])
 
# Adding title
plt.title("Credit Score <550")
 
# Removing top axes and right axes
# ticks
ax.get_xaxis().tick_bottom()
ax.get_yaxis().tick_left()
     
# show plot
plt.show()
```

```python
import seaborn as sns
sns.lineplot(y = CS_A['>-0.75 and <=-0.5'], x = CS_A['Date'])
sns.lineplot(y = CS_A['>-1 and <=-0.75'], x = CS_A['Date'])
sns.lineplot(y = CS_A['>-1.25 and <=-1'], x = CS_A['Date'])
plt.title("Credit Score 550> and <=680")
plt.legend(labels = ['>-0.75 and <=-0.5', '>-1 and <=-0.75',
                    '>-1.25 and <=-1'])
plt.gcf().set_size_inches(15, 8)
plt.xticks(CS_A['Date'][::6], rotation = 90);
```
